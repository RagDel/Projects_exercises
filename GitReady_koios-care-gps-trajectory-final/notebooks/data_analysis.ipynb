```python
# Attempting a different approach to parse and extract walking_time from 'data_dump' column
df = user_2_15km.copy()
walking_times = []
dates = []
for dump in df['data_dump']:
    for data in dump:
        date = list(data.keys())[0]
        date = pd.to_datetime(date).date()  # Converts to datetime and then to date
        dates.append(date)
        walking_times.append(data[list(data.keys())[0]]['time_away_from_home_per_day'])

walking_seconds = [time.total_seconds()/3600 for time in walking_times]


averaged_walking_seconds = [sum(walking_seconds[i:i+5]) / 5 for i in range(0, len(walking_seconds), 1)]

# Calculate midpoint dates for each group of three
midpoint_dates = []
for i in range(0, len(dates), 1):
    if i+4 < len(dates):
        midpoint = dates[i] + (dates[i+4] - dates[i]) // 4
        midpoint_dates.append(midpoint)
        
# Plotting the walking times
plt.figure(figsize=(10, 6))
plt.plot(midpoint_dates, averaged_walking_seconds[:-1], marker=',', linestyle='-')
plt.title('Average Time Away From Home Over 5-Day Periods')
plt.xlabel('Date')
plt.ylabel('Average Time away from home (hours)')
plt.xticks(rotation=45)
#plt.grid(True)
plt.show()
```
